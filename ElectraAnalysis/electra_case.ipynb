```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation Parameters
n_simulations = 10000

# Probabilities for each scenario (these can be adjusted based on assumptions)
prob_best_case = 0.2
prob_moderate_case = 0.65
prob_worst_case = 0.15


# Revenue multipliers for each scenario
# These multipliers represent the percentage growth in revenue under each scenario
revenue_multipliers = {
    'Option 1': {'best_case': 1.6, 'moderate_case': 1.2, 'worst_case': 0.8},
    'Option 2': {'best_case': 1.5, 'moderate_case': 1.1, 'worst_case': 0.5},
    'Option 3': {'best_case': 1.5, 'moderate_case': 1.25, 'worst_case': 0.8}
}

# Initial revenue from each region (in millions USD)
initial_revenue = {
    'Option 1': 5000,
    'Option 2': 3000,
    'Option 3': 2000
}


# Simulate revenue outcomes for each region
def simulate_revenue(region):
    outcomes = []
    for _ in range(n_simulations):
        scenario = np.random.choice(['best_case', 'moderate_case', 'worst_case'],
                                    p=[prob_best_case, prob_moderate_case, prob_worst_case])
        multiplier = revenue_multipliers[region][scenario]
        revenue = initial_revenue[region] * multiplier
        outcomes.append(revenue)
    return np.array(outcomes)

# Run simulations for each region
revenue_europe = simulate_revenue('Option 1')
revenue_asia = simulate_revenue('Option 2')
revenue_latam = simulate_revenue('Option 3')




# Calculate total revenue across all regions
total_revenue = revenue_europe + revenue_asia + revenue_latam



```

```python
plt.figure(figsize=(12, 6))

plt.hist(total_revenue, bins=10, color='green', edgecolor='black')
plt.title('Distribution of Added Revenue After Global Expansion')
plt.xlabel('Total Revenue (Millions USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Calculate and print key statistics
mean_revenue = np.mean(total_revenue)
median_revenue = np.median(total_revenue)
std_revenue = np.std(total_revenue)
worst_5th_percentile = np.percentile(total_revenue, 5)
best_95th_percentile = np.percentile(total_revenue, 95)

print(f"Mean Total Revenue: ${mean_revenue:.2f} million")
print(f"Median Total Revenue: ${median_revenue:.2f} million")
print(f"Standard Deviation of Total Revenue: ${std_revenue:.2f} million")
print(f"Worst 5th Percentile Revenue: ${worst_5th_percentile:.2f} million")
print(f"Best 95th Percentile Revenue: ${best_95th_percentile:.2f} million")
```
